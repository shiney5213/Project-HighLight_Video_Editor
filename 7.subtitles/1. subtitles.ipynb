```python
import pandas as pd
```

```python
datapath = 'substitle.txt'

with open(datapath, 'r', encoding ='UTF-8' ) as f:
    datas=f.read()
datas
```

```python
datas = datas.replace("'", '')
datas
```

```python
len(datas)
```

```python
datas = datas.split('\\n\\n')
datas
```

```python
sub1 = []
for data in datas:
    if data !='' :
        sub1.append(data.split('\\n'))
sub1
```

```python
len(sub1)
```

```python
sub2 =[]
for sub in  sub1:
    timedata =sub[1][:8]+'-->'+sub[1][18:26]
    print(timedata)
    sub2.append([sub[0], timedata, sub[2]])
sub2
```

```python
save_path = 'subtitle_final.txt'
with open(save_path, 'w', encoding = 'UTF-8') as f:
    f.write(str(sub2))
    
```

```python
context = {}
for (i, time, text) in sub2:
    context[i]=[time, text]
print(context)
```
